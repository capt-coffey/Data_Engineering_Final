```python
from pyspark.sql import functions as F
parcels_raw = spark.read.parquet("/Volumes/workspace/default/raw/parcels")


print(f"Total rows: {parcels_raw.count():,}")
print(f"\nSchema:")
parcels_raw.printSchema()

print(f"\nSample neighborhood values:")
display(
    parcels_raw.groupBy("neighborhood")
    .count()
    .orderBy("count", ascending=False)
    .limit(10)
)

print(f"\nVacancy distribution:")
display(
    parcels_raw.groupBy("is_vacant")
    .count()
    .orderBy("count", ascending=False)
)

print(f"\nAssessment value stats:")
display(
    parcels_raw.select(
        F.min("assessed_value").alias("min"),
        F.max("assessed_value").alias("max"),
        F.round(F.avg("assessed_value"), 2).alias("avg"),
    )
)
```
